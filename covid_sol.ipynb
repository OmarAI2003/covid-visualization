```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("covid.csv",index_col='Country/Region')
```

```python
df['total']=df.iloc[:,3:].sum(axis=1)
```

```python
df.head()
```

```python
top5=df.sort_values('total',ascending=False)[:5].iloc[:,3:-1]
```

```python
top5.T.plot()
plt.xlabel('Date')
plt.ylabel('Count of deaths')
```
